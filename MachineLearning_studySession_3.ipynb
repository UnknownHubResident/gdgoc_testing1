```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

```python
pd.set_option('display.max_columns', None)
IndoQuacke2022 = pd.read_csv('/content/katalog_gempa_2022.csv')
IndoQuacke2022.head()
```

```python
IndoQuacke2022.info()
```

```python
#filtered table to find the magnitude scale of earthquake in March 2022
MarchMagnitude2022 = IndoQuacke2022[IndoQuacke2022['date of event'].str.contains('March')]
MarchMagnitude2022.head()
```

```python
#earthquake magnitude variation in the 1st of march 2022
BarChartX = MarchMagnitude2022['date of event']
BarChartY = MarchMagnitude2022['magnitude']

plt.figure(figsize=(6, 6))
plt.title('earthquake magnitude variation in March 2022', fontsize = 24)
plt.xlabel('date of event', fontsize = 16)
plt.ylabel('magnitude', fontsize = 16)
plt.bar(BarChartX, BarChartY)
plt.show()
```

```python
plt.figure(figsize=(24, 16))

# Correctly pass X and Y values separately
plt.scatter(IndoQuacke2022['latitude (degree)'], IndoQuacke2022['longitude (degree)'])

# Formatting the plot
plt.title("Earthquake Points in Indonesia (2022)", fontsize=24, fontweight='bold')
plt.xlabel('Latitude (degree)')
plt.ylabel('Longitude (degree)')

plt.show()
```
